```python
%matplotlib inline

#add necessary libraries
import pycombo # combo community detection package
import matplotlib.pyplot as plt #plotting
import geopandas as gpd
from scipy import optimize
%pylab inline 
import os
from networkx.algorithms import community
import warnings
warnings.filterwarnings('ignore')
```

```python
community = gpd.read_file('Boundaries - Community Areas (current).geojson').set_index('community')
community['centroid'] = community.geometry.centroid
community.head()
```

```python
data.head()
```

```python
data.columns.unique()
```

```python
data_=data
data_['StartCommunityAreaNumber']=data_['Start Community Area Number']
data_['StartCommunityAreaNumber']=data_.StartCommunityAreaNumber.astype(int)
data_['StartCommunityAreaNumber']
```

```python
data_['EndCommunityAreaNumber']=data_['End Community Area Number']
data_['EndCommunityAreaNumber']=data_.EndCommunityAreaNumber.astype(int)
data_['EndCommunityAreaNumber']
```

```python
data_['TripDistance']=data_['Trip Distance']
data_['TripDistance']=data_.TripDistance.astype(int)
```

```python
data_['TripDuration']=data_['Trip Duration']
data_['TripDuration']=data_.TripDistance.astype(int)
```

```python
count_start=data_[['TripDistance','StartCommunityAreaNumber']].groupby(by='StartCommunityAreaNumber').count()
start=data_[['TripDistance','TripDuration','StartCommunityAreaNumber']].groupby(by='StartCommunityAreaNumber').sum()
start['trip_count']=count_start['TripDistance']
start
```

```python
count_end=data_[['TripDistance','EndCommunityAreaNumber']].groupby(by='EndCommunityAreaNumber').count()
end=data_[['TripDistance','TripDuration','EndCommunityAreaNumber']].groupby(by='EndCommunityAreaNumber').sum()
end['trip_count']=count_end['TripDistance']
end
```

```python
community.geometry.boundary.plot(figsize=(10,10), alpha=0.3)
```

```python
community['area_num_1']=community.area_num_1.astype(int)
```

```python
startGEO=community.merge(start,left_on='area_num_1',right_on='StartCommunityAreaNumber',how='right')
startGEO
```

```python
endGEO=community.merge(end,left_on='area_num_1',right_on='EndCommunityAreaNumber',how='right')
endGEO
```

```python
'''plot the spatial distribution of Trip Distance'''
fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(20,8))
startGEO.to_crs({'init': 'epsg:2263'}).plot(ax=ax[0],column='TripDistance',
                                            colormap='Spectral_r',legend=True,
                                            missing_kwds={
                                                            "color": "lightgrey",
                                                            "edgecolor": "red",
                                                            "hatch": "///",
                                                            "label": "Missing values",
                                                            },)
ax[0].set_aspect('equal')
ax[0].set_title('Distribution of Trip Distance of E-Scooter in Chicago, CRS: EPSG 2263',fontsize=12) 
endGEO.to_crs({'init': 'epsg:2263'}).plot(ax=ax[1],column='TripDistance',
                                            colormap='Spectral_r',legend=True,
                                            missing_kwds={
                                                            "color": "lightgrey",
                                                            "edgecolor": "red",
                                                            "hatch": "///",
                                                            "label": "Missing values",
                                                            },)
ax[1].set_title('Distribution of Trip Distance of E-Scooter in Chicago, CRS: EPSG 2263',fontsize=12)
```
